```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_excel('FN_Training_Set.xlsx')
```

```python
train, test = train_test_split(data, test_size=0.1, random_state=999)
```

```python
data
```

```python
clean = train.dropna(subset=['Content'])
```

```python
tfidf = TfidfVectorizer(ngram_range=(1,3), min_df=2)
tfidf.fit(clean['Content'])
```

```python
tfidf.vocabulary_
```

```python
X = tfidf.transform(clean['Content Title'])
```

```python
clf = LinearSVC()
clf.fit(X, clean['fake_news_score'])
```

```python
test_X = tfidf.transform(test['Content Title'])
clf.score(test_X, test['fake_news_score'])
```

```python
print(clf.classes_)
print(len(clf.classes_))
```

```python
np.sum(clf.coef_ == 0, axis = 1)
```

```python
len(clf.coef_[0])
```

```python
print('coefficients:', clf.coef_.shape)
# print('length of vocabulary:', len(keys))
```

```python
ngram_map = {idx: ngram for ngram, idx in tfidf.vocabulary_.items()}
```

```python
ngram_map
```

```python
coeff = clf.coef_[0] # Last 3 features are from Features Union

idx = np.argsort(coeff)

n = 50
bottom_indecies = idx[:n]
top_indecies = idx[-n:]

bottom_ = [ngram_map[k] + " " + str(round(coeff[k],2)) for k in bottom_indecies]
top_ = [ngram_map[k] + " " + str(round(coeff[k],2)) for k in top_indecies]
#     print(cls)
#     print(top_[-5:][::-1])
#     print(bottom_[:5])
```
